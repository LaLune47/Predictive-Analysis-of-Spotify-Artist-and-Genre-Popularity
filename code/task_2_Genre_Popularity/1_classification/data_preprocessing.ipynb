### Dataset

```python
import pandas as pd

df = pd.read_csv('../../data/clustered_df.csv')
centroids_df = pd.read_csv('../../data/centroids_df.csv')
```

### Further Data Preprocessing
<ul>
    <li>Lag feature construction </li>
    <ul>
        <li>Sum genre popularity score over past year</li>
        <li>Average popularity scores of songs in a genre</li>
</ul>

```python
df_date_fmt = df.copy()
df_date_fmt['Date'] = pd.to_datetime(df['Date'])
df_date_fmt['Previous_Date'] = pd.to_datetime(df['Previous_Date'])
df_date_fmt
```

```python
result = []

for genre_idx, genre_scores in df_date_fmt.groupby('Predicted_Genre'):
    genre_scores = genre_scores.groupby('Date')['Points (Total)'].mean()
    genre_scores['Genre_Year_Score'] = genre_scores.rolling('365D').sum()
    result.extend([(genre_idx, date, score) for date, score in zip(genre_scores.index, genre_scores['Genre_Year_Score'].values)])

score_df = pd.DataFrame(result, columns=['Predicted_Genre', 'Previous_Date', 'Genre_Year_Score'])
score_df
```

```python
lag_feat_df = pd.merge(df_date_fmt, score_df, on=['Predicted_Genre', 'Previous_Date'], how='left').fillna(0)
lag_feat_df
```

- Replace track features with feature values of each genre

```python
genre_df = lag_feat_df.copy()[['Predicted_Genre', 'Genre_Year_Score', 'Points_Category']]
genre_df = genre_df.join(centroids_df, how='left', on='Predicted_Genre')
genre_df = genre_df.drop('Predicted_Genre', axis=1)
genre_df
```

- Train test split (80:20)

```python
from sklearn.model_selection import train_test_split

X = genre_df.drop('Points_Category', axis=1)
y = genre_df['Points_Category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.to_csv('../../data/genre_class_train_df.csv', index=False)
X_test.to_csv('../../data/genre_class_test_df.csv', index=False)
```

```python
import numpy as np

np.save('../../data/genre_class_train_y', y_train)
np.save('../../data/genre_class_test_y', y_test)
```

#### Prepare data for final genre propularity classification

```python
final_pred_df = pd.DataFrame()
final_pred_df['Predicted_Genre'] = df_date_fmt['Predicted_Genre'].unique()
final_pred_df['Previous_Date'] = df_date_fmt['Date'].max()
final_pred_df = pd.merge(final_pred_df, score_df, on=['Predicted_Genre', 'Previous_Date'], how='left')
final_pred_df = final_pred_df.join(centroids_df, how='left', on='Predicted_Genre')
final_pred_df
```

```python
final_pred_df.to_csv('../Data/genre_final_pred_df.csv', index=False)
```
