### Dataset

```python
import pandas as pd

df = pd.read_csv('data/Spotify_Dataset_V3.csv', delimiter=';')
df
```

### Data Preprocessing
#### 1. Date Format Conversion

```python
df_date_fmt = df.copy()
df_date_fmt['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df_date_fmt
```

#### 2. Lag Feature Construction
<ul>
    <li>Sum artist popularity score over past year</li>
    <li>Average popularity scores of artists (if more than 1 artist involved in the song)</li>
</ul>

```python
result = []

for artist_idx, artist_scores in df_date_fmt.groupby('Artist (Ind.)'):
    artist_scores = artist_scores.groupby('Date')['Points (Ind for each Artist/Nat)'].sum()
    artist_scores['Artist_Year_Score'] = artist_scores.rolling('365D').sum()
    result.extend([(artist_idx, date, score) for date, score in zip(artist_scores.index, artist_scores['Artist_Year_Score'].values)])

score_df = pd.DataFrame(result, columns=['Artist (Ind.)', 'Previous_Date', 'Artist_Year_Score'])
score_df
```

```python
df_date_fmt['Previous_Date'] = df_date_fmt['Date'] - pd.tseries.offsets.Day()
df_date_fmt
```

```python
lag_feat_df = pd.merge(df_date_fmt, score_df, on=['Artist (Ind.)', 'Previous_Date'], how='left').fillna(0)
mean_scores_df = lag_feat_df.groupby(['Title', 'id', 'Date', 'Artists', 'Points (Total)'])['Artist_Year_Score'].mean()
lag_feat_df = pd.merge(lag_feat_df.drop('Artist_Year_Score', axis=1), mean_scores_df, on=['Title', 'id', 'Date', 'Artists', 'Points (Total)'], how='left')
lag_feat_df
```

#### 3. Drop Duplicates
<br />
<i>
    Note: <br />
    <ul>
        <li>Same titles and same artists can get different IDs </li>
        <li>Different titles or same titles with different artists can get same IDs </li>
        <li>There are days with some missing ranks (e.g. 20-07-2017 does not have data for rank 5) </li>
        <li>Dropping duplicates based on 5 columns, i.e. Title, ID, Date, Artists and Points, gives all possible combinations of songs </li>
    </ul><br />
    Output: <br />
    - More rows than intended (470436 > 467200), due to same titles different artists, that are actually the same songs but cannot be detected, because there are cases where two songs have the same titles but are indeed from different artists and are two entirely different songs
</i>

```python
lag_feat_df = lag_feat_df.drop_duplicates(['Title', 'id', 'Date', 'Artists', 'Points (Total)'])
lag_feat_df
```

#### 4. Convert Popularity Scores to Categories
- Bin values into 5 categories

```python
import math
import numpy as np

cat_df = lag_feat_df.copy()
cat_df['Points_Category'] = cat_df['Points (Total)'].apply(lambda x: math.ceil((x / 200) * 5) if x > 0 else 1)
cat_df
```

#### 5. Train, Validation, Test Split
- Train set: 80%
- Test set: 20%

```python
from sklearn.model_selection import train_test_split

X_sup = cat_df.drop(['Rank', 'Title', 'Artists', 'Date', '# of Artist', 'Artist (Ind.)', '# of Nationality', 'Nationality', 
                     'Continent', 'Points (Total)', 'Points (Ind for each Artist/Nat)', 'id', 'Song URL', 'Points_Category',
                     'Previous_Date'], 
                axis=1)
X_sup
```

```python
y_class = cat_df['Points_Category'].values
```

```python
X_class_train, X_class_test, y_class_train, y_class_test = train_test_split(X_sup, y_class, test_size=0.20, random_state=42)

print("Classification: ")
print(f"Train Set: x - {X_class_train.shape}; y - {y_class_train.shape}")
print(f"Test Set: x - {X_class_test.shape}; y - {y_class_test.shape}")
```

```python
train_eda_df = cat_df.loc[X_class_train.index]
train_eda_df.shape
```

```python
genre_train = cat_df.loc[X_class_train.index]
genre_test = cat_df.loc[X_class_test.index]

print("Genre Popularity Prediction: ")
print(f"Train Set: {genre_train.shape}")
print(f"Test Set: {genre_test.shape}")
```

#### 6. Z-score Normalisation

```python
from sklearn.preprocessing import StandardScaler

class_scaler = StandardScaler().fit(X_class_train)
scaled_X_class_train = class_scaler.transform(X_class_train)
scaled_X_class_test = class_scaler.transform(X_class_test)
```

```python
track_features = ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Valence']

genre_scaler = StandardScaler().fit(genre_train[track_features])
genre_train[track_features] = genre_scaler.transform(genre_train[track_features])
genre_test[track_features] = genre_scaler.transform(genre_test[track_features])
```

```python
from pickle import dump

dump(genre_scaler, open('genre_scaler.pkl', 'wb'))
```

#### 7. Dimensionality Reduction 
Principal Component Analysis

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(scaled_X_class_train)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Cumulative Explained Variance', y=1.05)
plt.show()
```

```python
pca_85 = PCA(n_components=0.85)
class_train_pca_df = pca_85.fit_transform(scaled_X_class_train)
class_test_pca_df = pca_85.transform(scaled_X_class_test)
```

```python
print(f"Number of components: {pca_85.n_components_}")
print(f"Train Set PCA: {class_train_pca_df.shape}")
print(f"Test Set PCA: {class_test_pca_df.shape}")
```
